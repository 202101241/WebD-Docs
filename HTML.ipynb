##7) Image Element
* Used for showing image in our website.
* Syntax :

```html
    <img src="url" alt="Alternative Text Description" />
```

* Here `alt` Attribute is used for visually impaired persons to read about image description using website reader.
* It is not going to show anything on website.
* Following website can be used for generating random pictures for our project.

```html
    <img src="https://picsum.photos/200" alt="Random Image" />
```

##8) HTML Boilerplate
* This is Complete format of how a HTML file is created.

```html
<!DOCTYPE html>
<html lang="en">
	<head>
		<meta charset="UTF-8" />
		<title>My Website</title>
	</head>

	<body>
		<h1>Hello World!</h1>
	</body>
</html>
```
* In head part all of the website related settings are written. Not showing any content in website.
* In `<head>` we have `<title>` where the title of our website is written.
* Everything we are going to write in `<body> </body>` part.
